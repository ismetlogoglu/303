```python
# Read data from a text file
with open('data3.txt', 'r') as file:
    data3 = [float(line.strip()) for line in file]
```

```python
import numpy as np
from scipy.stats import gamma
from scipy.optimize import minimize

# Random sample from data3
np.random.seed(1982)
random_sample = np.random.choice(data3, 100, replace=False)

# Method of Moments Estimation
mean_x = np.mean(random_sample)
var_x = np.var(random_sample, ddof=1)

alpha_mme = mean_x**2 / var_x
beta_mme = var_x / mean_x

print("MME Alpha:", alpha_mme)
print("MME Beta:", beta_mme)

# Maximum Likelihood Estimation
def negative_log_likelihood(params):
    alpha, beta = params
    return -np.sum(gamma.logpdf(random_sample, a=alpha, scale=1/beta))

initial_guess = [alpha_mme, beta_mme]
result = minimize(negative_log_likelihood, initial_guess, bounds=[(0.001, None), (0.001, None)])
alpha_mle, beta_mle = result.x

print("MLE Alpha:", alpha_mle)
print("MLE Beta:", beta_mle)



```
